```python
img_pieces = [
    header_img(),
    items_img(),
    total_img(),
    footer_img(),
]
```

```python
img_pieces_height = sum(map(lambda i: i.height, img_pieces))
```

```python
new_im = Image.new('RGB', (WIDTH, img_pieces_height))
```
