```python
import imgkit
import json
import random
```

```python
with open('html/1.html', 'r') as html_file:
    html_original = html_file.read()
```

```python
item_names = [
    "ЧІКЕН МАКНАГЕТС9",
    "ЧІКЕН МАКНАГЕТС6",
    "ЧІКЕН МАКНАГЕТС3",
    "БІГМАК", 
    "ДАБЛ ЧІЗ", 
    "СОУС СИРНИЙ",  
    "КУТЧУП",
    "КАРТОПЛЯНІ ДІПИ",
    "КАРТОПЛЯ ФРІ",
    "КОКА-КОЛА МАЛА",
    "КОКА-КОЛА СЕРЕД",
    "КОКА-КОЛА ВЕЛИК",
    "СПРАЙТ МАЛИЙ",
    "СПРАЙТ СЕРЕД",
    "СПРАЙТ ВЕЛИК",
    "ФАНТА МАЛИЙ",
    "ФАНТА СЕРЕД",
    "ФАНТА ВЕЛИК",
    "ЧАЙ З БАРГАМОТОМ",
]
item_names_prefixes = ['', ' ', ' * 1 ', ' * 2 ', ' * 3 ']
item_names_sufixes = ['', ' МЕН', ' МЕНЮ']
```

```python
generated_item_names = []
for item in item_names:
    for p in item_names_prefixes:
        for s in item_names_sufixes:
            generated_item_names.append(p + item + s)
```

```python
def price():
    if random.random() > 0.5:
        return None
    
    number = str(random.randint(1, 500)) + '.' + str(random.randint(0, 9)) + str(random.randint(0, 9))
    
    if random.random() > 0.4:
        return number + ' A'
    
    if random.random() > 0.7:
        return number + ' ГРН'
    
    if random.random() > 0.7:
        return number + ' Г'
    
    return number
```

```python
def random_items(items_names, price_func, max_n = 15):
    res = []
    n = random.randint(1, max_n - 1)
    for i in range(n):
        res.append({'name': items_names[random.randint(0, len(items_names) - 1)], 'price': price_func()})
        
    return res
```

```python
for i in range(100):
    items = random_items(generated_item_names, price)
    for label_type in [None, 'item', 'item-name', 'item-price', 'total']:
        html_config = {'items': items, 'labels': label_type}
        html = html_original.replace('/*data*/', 'data = ' + json.dumps(html_config))
        label_type_name = label_type if label_type is not None else 'origin'
        imgkit.from_string(html, 'generated/' + str(i) + '_' + label_type_name + '.png', {"xvfb": "", 'width': 320, 'height': 864})
```
